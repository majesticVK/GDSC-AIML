## Linear regression model for Energy Consumption vs Square **Footage**

```python
from sklearn.linear_model import LinearRegression
energy_consumption_vs_square_footage_model = LinearRegression()
x = df[['Square_Footage']]
y = df['Energy_Consumption']
energy_consumption_vs_square_footage_model.fit(x,y)
```

```python
energy_consumption_vs_square_footage_model.intercept_
```

```python
energy_consumption_vs_square_footage_model.coef_
```

```python
predicted_data = energy_consumption_vs_square_footage_model.predict(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (12, 6))
sns.scatterplot(x = 'Energy_Consumption', y = 'Square_Footage', data = df, hue=df['Building_Type'], label = 'Actual Data')
plt.plot(x, predicted_data, color = 'red', )

plt.title("Energy Consumption vs Square Footage")
plt.xlabel("Energy Consumption")
plt.ylabel("Square Footage")
plt.legend()
plt.show()

plt.title("Energy Consumption vs Square Footage")
plt.xlabel("Energy Consumption")
plt.ylabel("Square Footage")
plt.legend()
plt.show()
```
